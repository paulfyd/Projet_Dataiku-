```python
#Test dataset
data_test = pd.read_csv('census_income_test.csv')
data_test.head(2)
```

```python
#To print the confusion matrix

# Code from scikit-learn
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()

classes_name=["-50k","+50k"]
```

## Data cleaning

First of all, I'm deleting the instance weight column, unuseful according to the metadata. 

```python
data_train.drop(columns = ['instance weight'], inplace=True)
data_test.drop(columns = ['instance weight'], inplace =True)
```

Let's see if there are duplicates. 

```python
print("There are",data_train.duplicated().sum(),"duplicates in the training set.")
print("There are",data_test.duplicated().sum(),"duplictaes in the test set.")
```

I drop the duplicates of the training set, to make sure no sample is more imporant than others.

```python
data_train = data_train.drop_duplicates()
```

Then, I must deal with the NaN value. Our dataset contains a lot of value "?" or "Not in the universe" that I consider as NaN values.

Let's do the replacement. 

```python
data_train = data_train.replace({ " ?": np.nan , " Not in universe" : np.nan})
data_train.head(2)
```

Now, I will look at the weight of these different NaN values for each feature, and delete the corresponding column if it contains too many NaN values. 

```python
print("Let's see the proportion of NaN values per feature")
print("------------------------------------------------------")
print((data_train.isnull().sum()/len(data_train)*100))
```

I set the suppression threshold at 49%, which eliminates the three migration columns, which are redundant and contain a lot of '?'.

```python
for col in data_train.columns:
    if data_train[col].isnull().sum()/(len(data_train))>0.49:
        print("Suppression of the feature :",col)
        data_train.drop(col,axis=1,inplace=True)
        data_test.drop(col,axis=1,inplace=True)
```

```python
print("The new number of features is",data_train.shape[1],'.')
```

No need to worry about the latest NaN values, which will simply be ignored during One-Hot-encoding, beacause they concern only categorical features. 

Now, I binarize the problem. 

```python
data_train['result'].replace(' - 50000.',0, inplace = True)
data_test['result'].replace(' - 50000.',0, inplace = True)
data_train['result'].replace(' 50000+.',1 , inplace = True)
data_test['result'].replace(' 50000+.',1 , inplace = True)
```

Eventually, let's see the correlations between the different continuous features, to see if there are redundancies. 

```python
#correlation matrix
plt.figure(figsize=(15,10))
sns.heatmap(data_train.drop(['result'],axis=1).corr(),annot=True)
```

There is not too much correlation, so there is no need to remove features here. 

## Data preprocessing

```python
y_train = data_train['result']
y_test = data_test['result']
X_train = data_train.drop('result',axis=1)
X_test = data_test.drop('result',axis=1)
```

I use the one-hot encoding method to deal with the categorical features. It ignores the remaining NaN values. 

```python
X = pd.concat([X_train, X_test], axis = 0)
X = pd.get_dummies(X)
```

```python
X_train = X[:data_train.shape[0]]
X_test = X[data_train.shape[0]:]
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

I scale the data to make sure each feature has the same importance. 

```python
scaler = preprocessing.StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Feature selection 

In total I have 298 features. When it comes to features selection, I decided that I wanted to make the best compromises between :
- keeping a lot of features, in order to have all the information and reaching the higher performance for my model.
- dropping features, in order to avoid overfitting and obtain a good interpretability of the results.

I decided to work with 3 datasets, with 298, 150 and 80 features. To perform features selection, I used a filter based methode, that uses the Pearson correlation. It keeps the 150/80 features that are the more correlated to the target variable.

```python
# Lookin for the best 150 features using the correlation 

num_feats = 150
feature_name = X_train.columns.tolist()

cor_list = [] 

for i in X_train.columns.tolist():
    cor = np.corrcoef(X_train[i].values, y_train)[0,1]
    cor_list.append(cor)
    
# replace NaN with 0
cor_list = [0 if np.isnan(i) else i for i in cor_list]

# feature name
cor_feature = X_train.iloc[:,np.argsort(np.abs(cor_list))[-num_feats:]].columns.tolist()

# feature selection? 0 for not select, 1 for select
cor_support = [True if i in cor_feature else False for i in feature_name]

cor_feature_pd = pd.DataFrame(cor_feature)
cor_feature_pd
```

```python
features_kept_names = cor_feature_pd.values

X_train150 = X_train[features_kept_names.flatten()]
X_test150 = X_test[features_kept_names.flatten()]

print("The size of the new dataset is", X_train150.shape)
```

```python
scaler = preprocessing.StandardScaler()
X_train150_scaled = scaler.fit_transform(X_train150)
X_test150_scaled = scaler.transform(X_test150)
```

```python
#Lookin for the 80 best features using the correlation 

num_feats = 80
feature_name = X_train.columns.tolist()
cor_list = [] 

for i in X_train.columns.tolist():
    cor = np.corrcoef(X_train[i].values, y_train)[0,1]
    cor_list.append(cor)
    
# replace NaN with 0
cor_list = [0 if np.isnan(i) else i for i in cor_list]

# feature name
cor_feature = X_train.iloc[:,np.argsort(np.abs(cor_list))[-num_feats:]].columns.tolist()

# feature selection? 0 for not select, 1 for select
cor_support = [True if i in cor_feature else False for i in feature_name]

cor_feature_pd = pd.DataFrame(cor_feature)
cor_feature_pd
```

```python
features_kept_names = cor_feature_pd.values

X_train80 = X_train[features_kept_names.flatten()]
X_test80 = X_test[features_kept_names.flatten()]

print("The size of the new dataset is", X_train80.shape)
```

```python
scaler = preprocessing.StandardScaler()
X_train80_scaled = scaler.fit_transform(X_train80)
X_test80_scaled = scaler.transform(X_test80)
```

Now I have three different dataset, ready to be trained by different models. 

## Testing different models

### Premier modÃ¨le

Un premier modÃ¨le test : on va faire du linÃ©aire parce que blablabla

```python
ridge = RidgeClassifier()
ridge.fit(X_train80_scaled,y_train)

#Confusion matrix to analyse the faults 
y_pred=ridge.predict(X_test80_scaled)
y_train_pred=ridge.predict(X_train80_scaled)

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
cnf_matrix_train = confusion_matrix(y_train, y_train_pred)

# Plot normalized confusion matrix
plt.figure(figsize=(12,7))
plt.subplot(121)
plot_confusion_matrix(cnf_matrix_train, classes= classes_name ,normalize=True)
plt.title("Confusion matrix : Ridge (training dataset)")


plt.subplot(122)
plot_confusion_matrix(cnf_matrix, classes= classes_name ,normalize=True)
plt.title("Confusion matrix : Ridge (test dataset)")
plt.show()


print("The precision obtained is",(1-np.abs(y_pred-y_test).sum()/len(y_pred))*100)
print("The F1-score obtained is", f1_score(y_pred,y_test))
```

On a first model, linear, supposed to work well when there are a lot of features like here, we can see that the precision is very good. There is no overfitting, the results are the same on the training set and the test set. On the other hand, the model has a lot of difficulty to recognize people who earn more than 50k. 
That's why the F1-score is very low (0.27). Our goal will be to maximize the F1-score. Indeed, maximizing the precision is not very interesting as the classes are very unbalanced. A model which would return 0 systematically would have a good precision, and yet, it would not have learned anything...

Let's try with another dataset with more feature. 

```python
ridge = RidgeClassifier()
ridge.fit(X_train_scaled,y_train)

#Confusion matrix to analyse the faults 
y_pred=ridge.predict(X_test_scaled)
y_train_pred=ridge.predict(X_train_scaled)

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
cnf_matrix_train = confusion_matrix(y_train, y_train_pred)

# Plot normalized confusion matrix
plt.figure(figsize=(12,7))
plt.subplot(121)
plot_confusion_matrix(cnf_matrix_train, classes= classes_name ,normalize=True)
plt.title("Confusion matrix : Ridge (training dataset)")


plt.subplot(122)
plot_confusion_matrix(cnf_matrix, classes= classes_name ,normalize=True)
plt.title("Confusion matrix : Ridge (test dataset)")
plt.show()


print("The precision obtained is",(1-np.abs(y_pred-y_test).sum()/len(y_pred))*100)
print("The F1-score obtained is", f1_score(y_pred,y_test))
```

The result is a little better, but the f1-score remains very low. We won't try to tune the regularization parameter alpha for the moment, because there is no overfitting. 
On the other hand, there is a parameter which is interesting in our case of unbalanced classes. This is the class_weight parameter. We will make sure that errors on class 1 (50k+) are more penalized than errors on class 0 (-50k). This should increase the f1-score and decrease the accuracy.

```python
print('There is',1/(y_train.sum()/len(y_train)),'times more people with -50k.')
```

So we're going to penalize 12 times more errors on the +50k.

```python
ridge = RidgeClassifier(class_weight={0:1,1:12})
ridge.fit(X_train_scaled,y_train)

#Confusion matrix to analyse the faults 
y_pred=ridge.predict(X_test_scaled)
y_train_pred=ridge.predict(X_train_scaled)

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
cnf_matrix_train = confusion_matrix(y_train, y_train_pred)

# Plot normalized confusion matrix
plt.figure(figsize=(12,7))
plt.subplot(121)
plot_confusion_matrix(cnf_matrix_train, classes= classes_name ,normalize=True)
plt.title("Confusion matrix : Ridge (training dataset)")
plt.subplot(122)
plot_confusion_matrix(cnf_matrix, classes= classes_name ,normalize=True)
plt.title("Confusion matrix : Ridge (test dataset)")
plt.show()


print("The precision obtained is",(1-np.abs(y_pred-y_test).sum()/len(y_pred))*100)
print("The F1-score obtained is", f1_score(y_pred,y_test))
```

```python
ridge = RidgeClassifier(class_weight={0:1,1:12})
ridge.fit(X_train150_scaled,y_train)

#Confusion matrix to analyse the faults 
y_pred=ridge.predict(X_test150_scaled)
y_train_pred=ridge.predict(X_train150_scaled)

print("The precision obtained with 150 features is",(1-np.abs(y_pred-y_test).sum()/len(y_pred))*100)
print("The F1-score obtained with 150 features is", f1_score(y_pred,y_test))
```
